```python
def treat_outlier(x):
    # taking 5,25,75 percentile of column
    q1=np.percentile(x,25)
    q3=np.percentile(x,75)
    # calculating IQR range
    IQR=q3-q1
    lower_bound=q1-(1.5*IQR)
    upper_bound=q3+(1.5*IQR)
    #capping outliters
    return x.apply(lambda y : upper_bound if y > upper_bound else y).apply(lambda y : lower_bound if y < lower_bound else y)
```

```python
for i in num_list:
    df[i]=treat_outlier(df[i])
```

```python
df_train_all = pd.concat([x_train_all,y_train], axis=1)
df_test_all = pd.concat([x_test_all,y_test], axis=1)
```

```python
print(df_train_all.shape)
print(df_test_all.shape)
```

```python
def formula_from_cols(df, y):
    return y + ' ~ ' + ' + '.join([col for col in df.columns if not col==y])
```

```python
label = 'price'
f_1 = formula_from_cols(df_train_all,label)
f_1
```

```python
lm_all = smf.ols(formula= f_1, data = df_train_all).fit()
```

```python
print(lm_all.summary())
```

Observation Adj.R-squared (Coefficient of determination) is at 87.4% i.e. model is able to explain 87.4% of variation in the data. Difference between R-squared and Adj.R-Squared is due to the inclusion of non-significant predictors in the model. There are 18 fields, 'Intercept', 'symboling', 'normalized_losses' , 'fuel_type' , 'aspiration' , 'num_of_doors' , 'body_style' , 'length' , 'width' , 'height' , 'curb_eight' , 'engine_type' , 'num_of_cylinders' , 'fuel_system' , 'bore' , 'compression_ratio' , 'peak_rpm' ,'highway_mpg' p-value > 0.05. Hence, these 18 fields are not playing a significant role in predicting the Sales. These can be dropped one by one to observe the change in value of Adj.R-squared.

```python
predicted_train_all=lm_all.predict(df_train_all.drop('price',axis=1))
RMSE_train_all = metrics.mean_squared_error(y_train,predicted_train_all, squared=False)
RMSE_train_all.round(2)
```

```python
predicted_test_all=lm_all.predict(df_test_all.drop('price',axis=1))
RMSE_test_all = metrics.mean_squared_error(y_test,predicted_test_all, squared= False)
RMSE_test_all.round(2)
```
